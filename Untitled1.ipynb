```python
# Passo 1: Importação das Bibliotecas Necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

# Passo 2: Carregar o Arquivo CSV (No Google Colab, você pode fazer o upload diretamente)
uploaded = files.upload()

# Assumindo que o arquivo CSV se chama 'ecommerce_estatistica.csv'
df = pd.read_csv('ecommerce_estatistica.csv')

# Passo 3: Análise de Dados
# Verificando as primeiras linhas do dataframe
df.head()

# Verificando informações gerais do dataframe (tipos de dados, valores nulos, etc.)
df.info()

# Verificando estatísticas descritivas
df.describe()

# Passo 4: Criação dos Gráficos

## 1. Gráfico de Histograma
# Analisando a distribuição de uma variável, por exemplo 'Nota'
plt.figure(figsize=(10, 6))
sns.histplot(df['Nota'], bins=20, kde=True, color='skyblue')
plt.title('Distribuição das Notas', fontsize=16)
plt.xlabel('Nota', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.show()

## 2. Gráfico de Dispersão
# Analisando a relação entre 'Nota' e 'Preço_MinMax'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Nota', y='Preço_MinMax', data=df, color='green')
plt.title('Relação entre Nota e Preço', fontsize=16)
plt.xlabel('Nota', fontsize=12)
plt.ylabel('Preço Min-Max', fontsize=12)
plt.show()

## 3. Mapa de Calor (Heatmap)
# Verificando a correlação entre as variáveis numéricas
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Mapa de Calor das Correlações', fontsize=16)
plt.show()

## 4. Gráfico de Barra
# Exibindo a distribuição das avaliações por 'Marca'
plt.figure(figsize=(10, 6))
sns.countplot(x='Marca', data=df, palette='Set2')
plt.title('Distribuição das Avaliações por Marca', fontsize=16)
plt.xlabel('Marca', fontsize=12)
plt.ylabel('Número de Avaliações', fontsize=12)
plt.xticks(rotation=45)
plt.show()

## 5. Gráfico de Pizza
# Exibindo a distribuição do Gênero
plt.figure(figsize=(8, 8))
gender_distribution = df['Gênero'].value_counts()
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
plt.title('Distribuição por Gênero', fontsize=16)
plt.show()

## 6. Gráfico de Densidade
# Visualizando a distribuição de 'Desconto'
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Desconto'], shade=True, color='orange')
plt.title('Distribuição de Desconto', fontsize=16)
plt.xlabel('Desconto', fontsize=12)
plt.ylabel('Densidade', fontsize=12)
plt.show()

## 7. Gráfico de Regressão
# Verificando a relação entre 'Nota' e 'Qtd_Vendidos_Cod'
plt.figure(figsize=(10, 6))
sns.regplot(x='Nota', y='Qtd_Vendidos_Cod', data=df, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Regressão entre Nota e Quantidade Vendida', fontsize=16)
plt.xlabel('Nota', fontsize=12)
plt.ylabel('Quantidade Vendida (Cod)', fontsize=12)
plt.show()

```
